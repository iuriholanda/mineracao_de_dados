
### item f: Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado.



```python
dominio_esperado = {'date_message': str,
 'id_member_anonymous': str,
 'id_group_anonymous': str,
 'media': str,
 'media_type': str,
 'media_url': str,
 'has_media': bool,
 'has_media_url': bool,
 'trava_zap': bool,
 'text_content_anonymous': str,
 'dataset_info_id': int,
 'date_system': str,
 'score_sentiment': float,
 'score_misinformation': float,
 'id_message': int,
 'message_type': str,
 'messenger': str,
 'media_name': str,
 'media_md5': str}
```

```python
fora_do_dominio = []

for coluna, tipo_esperado in dominio_esperado.items():
    for idx, valor in arquivo[coluna].items():
        if pd.notnull(valor) and not isinstance(valor, tipo_esperado):
            fora_do_dominio.append((idx, coluna, valor))

for linha in fora_do_dominio:
    print(f"Valor fora do domínio encontrado na linha {linha[0]}, coluna '{linha[1]}': {linha[2]}")

```

```python
arquivo.value_counts('media_type')
```

```python
# Conversão de datas
arquivo['date_message'] = pd.to_datetime(arquivo['date_message'], errors='coerce')
arquivo['date_system'] = pd.to_datetime(arquivo['date_system'], errors='coerce')

# Lista para armazenar violações
violacoes = []

# Domínios esperados para colunas categóricas -- tipo object
valores_esperados = {
    'media_type': {'image/jpg', 'url', 'video/mp4', 'application/vnd.android.package-archive', 'image/mp4', 'text', None, np.nan},
    'message_type': {'Texto', 'Imagem', 'Url', 'Video', 'Application', 'Audio', None, np.nan},
    'messenger': {'whatsapp', 'telegram', 'signal', None, np.nan},
}

# Loop de validação
for idx, row in arquivo.iterrows():
    # score_sentiment: deve estar entre -1 e 1
    s = row['score_sentiment']
    if pd.notnull(s) and (s < -1 or s > 1):
        violacoes.append((idx, 'score_sentiment', s))

    # score_misinformation: deve estar entre 0 e 1
    m = row['score_misinformation']
    if pd.notnull(m) and (m < 0 or m > 1):
        violacoes.append((idx, 'score_misinformation', m))

    # dataset_info_id: deve ser exatamente 5
    d = row['dataset_info_id']
    if d != 5:
        violacoes.append((idx, 'dataset_info_id', d))

    # id_message: deve ser inteiro positivo
    im = row['id_message']
    if not isinstance(im, (int, np.integer)) or im < 0:
        violacoes.append((idx, 'id_message', im))

    # Datas válidas
    if pd.isnull(row['date_message']):
        violacoes.append((idx, 'date_message', 'data inválida'))

    if pd.isnull(row['date_system']):
        violacoes.append((idx, 'date_system', 'data inválida'))

    # Colunas booleanas
    for col in ['has_media', 'has_media_url', 'trava_zap']:
        if not isinstance(row[col], (bool, np.bool_)):
            violacoes.append((idx, col, row[col]))

    # Colunas categóricas com domínio definido
    for col, dominio in valores_esperados.items():
        val = row[col]
        if pd.notnull(val) and val not in dominio:
            violacoes.append((idx, col, val))

# Resultado
print(f"Total de violações de domínio: {len(violacoes)}")
for v in violacoes[:10]:
    print(f"Violação na linha {v[0]}, coluna '{v[1]}': valor '{v[2]}'")
```

### item i: Crie uma coluna chamada “viral” contendo o valor 0 se o texto da mensagem (valor do campo “text”) não for encontrado em outras linhas e 1, caso contrário.

```python
arquivo['viral'] = (arquivo.duplicated(subset='text_content_anonymous', keep = False) & arquivo['text_content_anonymous'].notna()).astype(int)
arquivo['viral'][1]
```

###item j:  Crie uma coluna chamada “sharings” contendo a quantidade de vezes que o texto armazenado no atributo “text” aparece no dataset.

```python
arquivo['sharings'] = arquivo['text_content_anonymous'].map(arquivo['text_content_anonymous'].value_counts())
arquivo['sharings'][2]
```

```python
arquivo[arquivo['viral']==1].shape == arquivo[arquivo['sharings']>1].shape
```

###item k: Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos.

```python
def analise_sentimento(value):
  if value > 0.05:
    return 1
  if value < -0.05:
    return -1
  return 0

arquivo['sentiment'] = arquivo['score_sentiment'].apply(analise_sentimento)
arquivo.head()
```
